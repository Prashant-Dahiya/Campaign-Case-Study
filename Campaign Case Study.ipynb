```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("10.4-spend data for campaigns data file.xls")
df
```

```python
df.drop('Unnamed: 7', inplace=True, axis=1)
df
```

```python
df[['start date','end date']]=df['Duration'].str.split('-', expand=True)
df
```

```python
df['start day of week']=df['start date'].dt.day_of_week
df
# Monday=0 and Sunday=6
```

```python
df['end day of week']=df['end date'].dt.day_of_week
df
```

```python
df['friday date_add start']=0
for i in df['start day of week'].index:
    if df['start day of week'][i]==0:
        df['friday date_add start'][i]+=4
    elif df['start day of week'][i]==1:
        df['friday date_add start'][i]+=3
    elif df['start day of week'][i]==2:
        df['friday date_add start'][i]+=2
    elif df['start day of week'][i]==3:
        df['friday date_add start'][i]+=1
    elif df['start day of week'][i]==4:
        df['friday date_add start'][i]+=0
    elif df['start day of week'][i]==5:
        df['friday date_add start'][i]+=6
    elif df['start day of week'][i]==6:
        df['friday date_add start'][i]+=5
df
```

```python
df['friday date_add end']=0
for i in df['end day of week'].index:
    if df['end day of week'][i]==0:
        df['friday date_add end'][i]+=4
    elif df['end day of week'][i]==1:
        df['friday date_add end'][i]+=3
    elif df['end day of week'][i]==2:
        df['friday date_add end'][i]+=2
    elif df['end day of week'][i]==3:
        df['friday date_add end'][i]+=1
    elif df['end day of week'][i]==4:
        df['friday date_add end'][i]+=0
    elif df['end day of week'][i]==5:
        df['friday date_add end'][i]+=6
    elif df['end day of week'][i]==6:
        df['friday date_add end'][i]+=5
df
```

```python
# df['friday start']=df['start date']+pd.Timedelta(df['friday start date'])
df['friday start date'] = df['start date'] + pd.to_timedelta(df['friday date_add start'], unit='D')
df
```

```python
# df=df.rename({'friday start date':'friday date_add start'}, axis=1)
# df
# df.drop('friday start',axis=1,inplace=True)
```

```python
df['friday end date'] = df['end date'] + pd.to_timedelta(df['friday date_add end'], unit='D')
df
```

```python
df['duration']=df['end date']-df['start date']+pd.to_timedelta(1,unit='D') #add date syntax
df
```

```python
df=df.rename({'product ':'product'}, axis=1)
df
```

```python
df['no of products']=0

for i in df['product'].index:
    df['no of products'][i]=len(df['product'][i].split('-'))
df
```

```python
type(df['duration'][0])
```

```python
df['duration days']=df['duration'].dt.days
df
```

```python
df['day spend per product']=df['Spend']/(df['no of products']*df['duration days'])
df
```

```python
min_friday_start_date=df['friday start date'].min()
min_friday_start_date
```

```python
max_friday_end_date=df['friday end date'].max()
max_friday_end_date
```

```python
arr=[]
a=min_friday_start_date
arr.append(a)
for i in range(100):
    a=a+pd.to_timedelta(7,unit='D')
    arr.append(a)
    if arr[-1]==max_friday_end_date:
        break
    i+=1
arr
```

```python
wd=pd.Series(arr, name='weekdays')
wd
```
